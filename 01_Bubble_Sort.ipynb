```python
def meu_bubble_sort(lista):
    n = len(lista)

    # Percorre toda a lista
    for i in range(n):
        # Os últimos i elementos já estão no lugar, então não precisamos olhar
        for j in range(0, n - i - 1):

            # Compara o elemento atual com o próximo
            if lista[j] > lista[j + 1]:
                # Se estiverem na ordem errada, troca!
                lista[j], lista[j + 1] = lista[j + 1], lista[j]

    return lista

# Testando:
# Altere os números abaixo para modificar a lista.
numeros = [10, 5, 8, 2, 7, 1, 9]
print(f"Lista original: {numeros}")

lista_ordenada = meu_bubble_sort(numeros)
print(f"Lista ordenada: {lista_ordenada}")
```
