```python
class PilhaHistorico:
    def __init__(self):
        # Iniciamos uma lista vazia que servirá como nossa pilha
        self.pilha = []

    def adicionar_pagina(self, url):
        """Método Push: Adiciona uma nova página ao topo."""
        self.pilha.append(url)
        print(f"🌐 Visitou: {url}")

    def voltar_pagina(self):
        """Método Pop: Remove a página do topo e retorna para a anterior."""
        if not self.esta_vazia():
            removido = self.pilha.pop()
            print(f"⬅️ Saindo de: {removido}")
        else:
            print("⚠️ Histórico vazio! Não há para onde voltar.")

    def esta_vazia(self):
        """Verifica se a pilha está vazia."""
        return len(self.pilha) == 0

    def exibir_topo(self):
        """Mostra qual a página atual (o topo)."""
        if not self.esta_vazia():
            print(f"📍 Você está em: {self.pilha[-1]}")

# --- Testando a nossa Pilha ---
meu_navegador = PilhaHistorico()

meu_navegador.adicionar_pagina("google.com")
meu_navegador.adicionar_pagina("github.com")
meu_navegador.adicionar_pagina("colab.research.google.com")

meu_navegador.exibir_topo()
meu_navegador.voltar_pagina()
meu_navegador.exibir_topo()
```
