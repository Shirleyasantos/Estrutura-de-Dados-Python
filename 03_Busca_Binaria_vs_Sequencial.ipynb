```python
import time
import math

# 1. Busca Sequencial - O(N)
def busca_sequencial(lista, alvo):
    for i in range(len(lista)):
        if lista[i] == alvo:
            return i
    return -1

# 2. Busca Binária - O(log N)
def busca_binaria(lista, alvo):
    esquerda, direita = 0, len(lista) - 1
    while esquerda <= direita:
        meio = (esquerda + direita) // 2
        if lista[meio] == alvo:
            return meio
        elif lista[meio] < alvo:
            esquerda = meio + 1
        else:
            direita = meio - 1
    return -1

# --- O TESTE DE PERFORMANCE ---
# Criando uma lista gigante de 0 a 1.000.000
lista_gigante = list(range(1000000))
alvo = 999999  # Vamos buscar o último número (pior caso)

# Testando Busca Sequencial
inicio = time.time()
busca_sequencial(lista_gigante, alvo)
fim = time.time()
print(f"⏱️ Busca Sequencial O(N): {(fim - inicio):.5f} segundos")

# Testando Busca Binária
inicio = time.time()
busca_binaria(lista_gigante, alvo)
fim = time.time()
print(f"⚡ Busca Binária O(log N): {(fim - inicio):.5f} segundos")
```
