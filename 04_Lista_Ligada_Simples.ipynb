```python
class No:
    """Esta classe representa um único 'Nó' da nossa lista."""
    def __init__(self, dado):
        self.dado = dado
        self.proximo = None  # Por padrão, o próximo é vazio

class ListaLigada:
    """Esta classe gerencia a conexão entre os nós."""
    def __init__(self):
        self.cabeca = None  # A lista começa vazia

    def inserir_no_inicio(self, novo_dado):
        """Cria um nó e o coloca na frente de todos."""
        novo_no = No(novo_dado)
        novo_no.proximo = self.cabeca
        self.cabeca = novo_no
        print(f"✅ {novo_dado} inserido no início!")

    def exibir_lista(self):
        """Percorre a lista e imprime os elementos."""
        atual = self.cabeca
        elementos = []
        while atual:
            elementos.append(str(atual.dado))
            atual = atual.proximo
        print(" -> ".join(elementos) + " -> None")

    def inserir_no_final(self, novo_dado):
        """Cria um nó e o coloca na última posição da lista."""
        novo_no = No(novo_dado)

        # Caso 1: Se a lista estiver vazia, o novo nó vira a cabeça
        if self.cabeca is None:
            self.cabeca = novo_no
            print(f"✅ {novo_dado} era a lista vazia, agora é o primeiro!")
            return

        # Caso 2: Percorrer a lista até chegar ao último nó
        ultimo = self.cabeca
        while ultimo.proximo:  # Enquanto existir um "próximo", continue andando
            ultimo = ultimo.proximo

        # Agora que chegamos ao último, fazemos ele apontar para o novo nó
        ultimo.proximo = novo_no
        print(f"✅ {novo_dado} inserido no final!")

# --- Testando ---
minha_lista = ListaLigada()
minha_lista.inserir_no_inicio(10)
minha_lista.inserir_no_inicio(20)
minha_lista.inserir_no_inicio(30)

print("\nEstrutura da sua Lista Ligada:")
minha_lista.exibir_lista()
```

```python
# --- Testando a nova função ---
minha_lista.inserir_no_final(5)
minha_lista.inserir_no_final(1)

print("\nLista após inserções no final:")
minha_lista.exibir_lista()
```
