```python
class No:
    """Esta classe representa um único 'Nó' da nossa lista."""
    def __init__(self, dado):
        self.dado = dado
        self.proximo = None  # Por padrão, o próximo é vazio

class ListaLigada:
    """Esta classe gerencia a conexão entre os nós."""
    def __init__(self):
        self.cabeca = None  # A lista começa vazia

    def inserir_no_inicio(self, novo_dado):
        """Cria um nó e o coloca na frente de todos."""
        novo_no = No(novo_dado)
        novo_no.proximo = self.cabeca
        self.cabeca = novo_no
        print(f"✅ {novo_dado} inserido no início!")

    def exibir_lista(self):
        """Percorre a lista e imprime os elementos."""
        atual = self.cabeca
        elementos = []
        while atual:
            elementos.append(str(atual.dado))
            atual = atual.proximo
        print(" -> ".join(elementos) + " -> None")

# --- Testando ---
minha_lista = ListaLigada()
minha_lista.inserir_no_inicio(10)
minha_lista.inserir_no_inicio(20)
minha_lista.inserir_no_inicio(30)

print("\nEstrutura da sua Lista Ligada:")
minha_lista.exibir_lista()
```
